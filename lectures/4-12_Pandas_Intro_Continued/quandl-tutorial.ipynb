### NOTE: Explanation of data

Adjusted Close and Open: 

> An adjusted closing price is a stock's closing price on any given day of trading that has been amended to include any distributions and corporate actions that occurred at any time prior to the next day's open. The adjusted closing price is often used when examining historical returns or performing a detailed analysis on historical returns.

> Read more: Adjusted Closing Price https://www.investopedia.com/terms/a/adjusted_closing_price.asp#ixzz5COp0BCVq 

Split ratio:

> The ratio of shares outstanding compared to what was originally available before a stock split. Investors use split ratio information to determine how much the stock of a publicly-traded company has been split. The higher the split ratio is, the lower the share price is.

> Read more: http://www.businessdictionary.com/definition/split-ratio.html

# Tip: Memoizing Function Calls

See here: https://stackoverflow.com/questions/815110/is-there-a-decorator-to-simply-cache-function-return-values?utm_medium=organic&utm_source=google_rich_qa&utm_campaign=google_rich_qa

You can add a decorator that will cache the output of a function.

```python
import functools
```

```python
@functools.lru_cache(maxsize=None)
def get_AAPL():
    df = quandl.get_table('WIKI/PRICES', ticker='AAPL')
    return df
```

```python
# First run will be slow
df = get_AAPL()
```

```python
df.head()
```

```python
# Second run will be fast. It doesn't have to download the data again
df = get_AAPL()
```

```python
df.head()
```
